```python
import pyspark
import json
from pyspark.sql.session import SparkSession
from pyspark.context import SparkContext
sc = SparkContext("local","Miniproject2")
spark = SparkSession(sc)
```

```python
def get_champion_win(match_dict_text):
  """
  this function output every champion for the winning team for every match
  """
  champoins_list=[]
  match_dict=json.loads(match_dict_text)
  for i in   match_dict["info"]["participants"]:
      #it is a list of dictionaries , each dictoinary is a team player
      if i['win']==True:
        champoins_list.append(i["championName"])
  return champoins_list

def get_champion_pick(match_dict_text):
  """
  This function outputs every champion for both teams for every match

  """
  champoins_list=[]
  match_dict=json.loads(match_dict_text)
  for i in match_dict["info"]["participants"]:
    champoins_list.append(i["championName"])
  return champoins_list

def get_championsID_bans(match_dict_text):
  """
  this function outputs every champion id that is banned from both teams

  """
  championsID_list=[]
  match_dict=json.loads(match_dict_text)
  for team in match_dict["info"]["teams"]:
    for ban in team["bans"]:
      championsID_list.append(ban["championId"])
  return championsID_list

def get_itemsID_pick(match_dict_text):
  """
  this function outputs all the items Ids picked from both teams for every match
  """
  itemsID_list=[]
  match_dict=json.loads(match_dict_text)

  for i in match_dict["info"]["participants"]:
    itemsID_list.extend([i["item0"],i["item1"],i["item2"],i["item3"],i["item4"],i["item5"]])

  return itemsID_list

def get_itemsID_win(match_dict_text):
  """
  this function outputs the items selected from the winning team
  """
  itemsID_list=[]
  match_dict=json.loads(match_dict_text)
  for player in match_dict["info"]["participants"]:

    if player['win']==True:
      itemsID_list.extend([player['item0'],player['item1'],player['item2'],player['item3'],player['item4'],player['item5']])
  return itemsID_list

from itertools import combinations
def get_champions_synergies_win(match_dict):
  """
  this function outputs a combination of the champions for the winning team for every match

  """
  champions_synergies_list=[]
  champions_list=get_champion_win(match_dict)
  champions_combinations=combinations(champions_list,2)
  for i in champions_combinations:
    champions_synergies_list.append('**'.join(list(sorted(i))))
  return champions_synergies_list


def get_items_champions_synergies(match_dict_text):
  """
  this function outputs a combination for every item selected by one player and the champion selected by the same player for both teams for every match
  """
  items_champions_list=[]
  match_dict=json.loads(match_dict_text)
  for player in  match_dict["info"]["participants"]:
    champion=player['championName']
    items_champions_list.extend([champion+'**'+str(i) for i in [player['item0'],player['item1'],player['item2'],player['item3'],player['item4'],player['item5']]])
  return items_champions_list
def get_items_win_for_champion(match_dict_text):
    champion='Lulu'# and 'Lulu'
    itemsID_list=[]
    match_dict=json.loads(match_dict_text)
    for player in match_dict["info"]["participants"]:
      if player["championName"]==champion:
        if player['win']==True:
          itemsID_list.extend([player['item0'],player['item1'],player['item2'],player['item3'],player['item4'],player['item5']])
    return itemsID_list 
def get_items_pick_for_champion(match_dict_text):
    champion='Lulu'# and 'Lulu'
    itemsID_list=[]
    match_dict=json.loads(match_dict_text)
    for player in match_dict["info"]["participants"]:
      if player["championName"]==champion:
        itemsID_list.extend([player['item0'],player['item1'],player['item2'],player['item3'],player['item4'],player['item5']])
    return itemsID_list 

from itertools import product
def get_champions_vs_wins(match_dict_text):
  """
  This function is supposed to print all champions that have faced each others 
  it results a list of tuples and every tuple have two champions the first champion is from the winning team and the second from the losing team
  """
  champion_won=[]
  champion_lost=[]
  output_vs=[]
  match_dict=json.loads(match_dict_text)
  for i in   match_dict["info"]["participants"]:
      #it is a list of dictionaries , each dictoinary is a team player
      if i['win']==True:
        champion_won.append(i["championName"])
      elif i['win']==False:
        champion_lost.append(i['championName'])
  output_vs_list=list(product(champion_won,champion_lost))
  for i in output_vs_list:
    output_vs.append('**'.join(list(i)))
  return output_vs

def get_champions_vs_times(match_dict_text):
  """
  This function is supposed to print all champions that have faced each others 
  """
  champion_won=[]
  champion_lost=[]
  output_vs=[]
  match_dict=json.loads(match_dict_text)
  for i in   match_dict["info"]["participants"]:
      #it is a list of dictionaries , each dictoinary is a team player
      if i['win']==True:
        champion_won.append(i["championName"])
      elif i['win']==False:
        champion_lost.append(i['championName'])
  output_vs_list=list(product(champion_won,champion_lost))
  for i in output_vs_list:
    output_vs.append('**'.join(list(sorted(i))))
  return output_vs



```

```python

words=matches_data.flatMap(lambda line:myfunction(line))
wordcount=words.map(lambda key:(key,1)).reduceByKey(lambda key,value: key+value).sortBy(lambda x: x[1], ascending=False)
result = wordcount.collect()

with open(fileName,'w+') as file:
  file.write(
      '\n'.join(str(i) for i in result)
  )
```
